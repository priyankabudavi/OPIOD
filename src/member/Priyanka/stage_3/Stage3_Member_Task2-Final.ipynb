#### Formally state the Null and Alternative Hypothesis and Define the type of hypothesis and the thresholds

Hypothesis is supposition or proposed explanation made on the basis of limited evidence as a starting point for further investigation. Also, its necessary to test this to carry out the study of a sample to variables. Thus NUll hyporthesis come into picture. 

Null Hypothsis (H0) : It states that there is no significant difference between the specified population. 
Alternative Hypothesis (H1) : The alternative is opposite of null hypothesis stating that there is relation ship between the variables.

The types of hypothesis are - 
1. One tailed Test - performed to check if the sample mean is less than or greater than the population mean.
2. Two tailed test - performed to find the significant difference between the variables.

There are different threshold that will help us understand if a null hypothesis should be accepted or rejected like pvalue

if the pvalue is less than 0.05 then we reject the null hypothesis. 

There are other functions like chi-contigency to check relation between two variables that are independant. 
P-Paired Test to check the if the data of interest differs with two different categories like checking weight of a person before the disease attacked him and then checking the weight after recovering. The Null hypothesis here is the weight remains same before and after. 

With the example above we will try to check for other 5 variables.

#### I grouped the data by state and created a dataframe that gives the mean of each state for each variable. Using the mean we will compare with the deaths value to understand if the mean of the two samples are same. Here am performing the test of independance. 

```python
df_mean_variables = hypo_df.groupby(['State'])[['Norm_deaths' ,'Drug overdose deaths raw value', 'Premature death raw value', 'Injury deaths raw value', 'Adult smoking raw value' , 'Firearm fatalities raw value' , 'Insufficient sleep raw value']].mean().sort_values(by  = "Norm_deaths", ascending = False).reset_index()
```

```python
df_mean_variables
```

#### 1. Test for Injury deaths raw value : 
    
    H0: The mean of Norm deaths and Injury deaths are the same 
    H1 : The mean of samples is different.

```python
res = stats.ttest_ind(a= df_mean_variables['Injury deaths raw value'],
                b= df_mean_variables['Norm_deaths'],
                equal_var=False)
res
```

```python
if res[1] <= 0.05:
    print("Reject the Null Hypothesis as the pvalue is less than 0.05")
else:
    print("Accept the Null Hypothesis")
```

The null hypothesis is rejected as the the mean of norm deaths and that of injury deaths are not same. Thus we accept the alternative hypothesis.

### M2.2 Perform linear regression to discover patterns (25 pts)

#### Perform a linear regression between Normalized Mortality and Opiod_Dispensing_Rate

#### We perform linear regression to understand the dependancy of a variable with other independant variable. 

```python
#normalize the opiod dispensing rate 
super_df['Opiod_Dispensing_Rate_Norm'] = (super_df['Opiod_Dispensing_Rate']/super_df['Population'] * 100000).round()
```

```python
linear_data = super_df[['Opiod_Dispensing_Rate_Norm' , 'Norm_deaths']]
```

```python
linear_data
```

```python
# create a fitted model using smf.ols
lm = smf.ols(formula='Norm_deaths ~ Opiod_Dispensing_Rate_Norm', data=linear_data).fit()
```

```python
preds = lm.predict(linear_data)
```

```python
lm.params
```

```python
lm.summary()
```

#### Inference : 

* The independent variable is opid dispensing rate and the dependent variable is norm deaths. 
*  The R-squared value is 0.001 i.e with just one independent variable â accounts for roughly 1% of the variance in death rates 
* we see most of the deaths were when the  opiod dispensing rates was equal to 100 units. 
* From the above parameters we see that the slope is positive this means that with the increase in opiod dispensing rate,        the deaths also increased.
* We can interpret the coefficient of opiod dispensing rate as, every year the deaths for opiod dispensing rate is found to increase by 12%
 

## Multiple linear regression with 5 variables

```python
#renaming the column names for simplicity
new_df = super_df.rename(columns={"Drug overdose deaths raw value": "Drug_overdose_deaths_raw_value", "Injury deaths raw value": "Injury_deaths_raw_value", "Adult smoking raw value": "Adult_smoking_raw_value" ,"Firearm fatalities raw value": "Firearm_fatalities_raw_value" , "Insufficient sleep raw value" : "Insufficient_sleep_raw_value"})
```

#### We need to normalize the variable that we are comparing to population as Opiod_dispensing rate is normalised. Thus we normalize all the variables and then perform the multiple regression

```python
new_df['Opiod_Dispensing_Rate_Norm'] = (new_df['Opiod_Dispensing_Rate']/super_df['Population'] * 100000).round()
```

```python
new_df['Drug_overdose_deaths_raw_value_Norm'] = (new_df['Drug_overdose_deaths_raw_value']/super_df['Population'] * 100000).round()
new_df['Injury_deaths_raw_value_Norm'] = (new_df['Injury_deaths_raw_value']/super_df['Population'] * 100000).round()
new_df['Adult_smoking_raw_value_Norm'] = (new_df['Adult_smoking_raw_value']/super_df['Population'] * 100000).round()
new_df['Firearm_fatalities_raw_value_Norm'] = (new_df['Firearm_fatalities_raw_value']/super_df['Population'] * 100000).round()
new_df['Insufficient_sleep_raw_value_Norm'] = (new_df['Insufficient_sleep_raw_value']/super_df['Population'] * 100000).round()
```

```python
#filling the nan values with zero after normalization
new_df = new_df.fillna(0)
```

```python
# create a fitted model with all the features of interest
lm = smf.ols(formula='Opiod_Dispensing_Rate_Norm ~ Drug_overdose_deaths_raw_value_Norm + Injury_deaths_raw_value_Norm + Adult_smoking_raw_value_Norm + Firearm_fatalities_raw_value_Norm + Insufficient_sleep_raw_value_Norm', data=new_df).fit()

# print the coefficients
lm.params
```

```python
lm.summary()
```

#### Inference : 

* Here the opioid dispensing rate is the dependent variable, and the rest other variables are independent.
* Injury_deaths_raw_value is  0.325279 positively related , Firearm_fatalities_raw_value is -0.281828 are negatively related. So when the slope is positive if independant variable increases by value then the dependent variable will also increase.

* Adult_smoking_raw_value_Norm , Insufficient_sleep_raw_value_Norm are 0 so that means the slope is zeo and the variables are not related. 

* Drug_overdose_deaths_raw_value is 0.0778 this means the slope is positive and the hence the death increase with every 7.78% increase in drug overdose. 

* The R-squared is 0.146 for this model and with 5 independent variables â accounts for roughly 14% of the variance in opioid dispensing rates.

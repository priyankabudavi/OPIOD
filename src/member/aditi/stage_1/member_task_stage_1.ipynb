## Task M3.1 - Merge all the three datasets to create a super datafame

```python
# Importing pandas
import pandas as pd
```

### Dataset 1 :  Drug Overdose Dataset

```python
# Read text file into pandas DataFrame
df_drug_overdose = pd.read_csv("../../../../data/stage_1/Underlying Cause of Death-County-2019.txt", sep="\t")
df_drug_overdose.head
```

### Read Dataset 2 :  County Health Rankings

```python
# Read csv file into pandas DataFrame
df_county_health_ranking = pd.read_csv("../../../../data/stage_1/County_Health_Ranking.csv")
df_county_health_ranking.head
```

### Read Dataset 3 :  County Opioid Dispensing Rates

```python
# Read csv file into pandas DataFrame
df_county_opioid_dispensing_rates = pd.read_csv("../../../../data/stage_1/2019-Opioid_Rate.csv",
                                                encoding='unicode_escape')
df_county_opioid_dispensing_rates.head
```

### Merging first two datasets based on county fips column

```python
# Merging two  datasets df_drug_overdose and df_county_health_ranking using 'County code' and '5-digit FIPS Code'/reference[1]
df_merged_drug_overdose_county_health_ranking = pd.merge(df_drug_overdose, df_county_health_ranking, 
                                                         left_on='County Code', 
                                                         right_on='5-digit FIPS Code')
df_merged_drug_overdose_county_health_ranking.head
```

### Superdataset obtained by merging third datasets based on county fips column

```python
# Merging df_merged_drug_overdose_county_health_ranking with df_county_opioid_dispensing_rates using 5-digit FIPS Code andFIPS
df_superdataset = pd.merge(df_merged_drug_overdose_county_health_ranking, df_county_opioid_dispensing_rates, 
                                                         left_on='5-digit FIPS Code', 
                                                         right_on='FIPS')
print(df_superdataset)
```

### Exporting dataset to csv format

```python
df_superdataset.to_csv("../../../../data/stage_1/superDatasetAditi.csv")
```

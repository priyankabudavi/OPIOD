#### M1.1 Shows comparisions between a variable and Normalized Mortality rate in a scatter plot -- Incorporate your best model prediction trend line - Linear / Non-Linear. (20 pts)

```python
import pandas as pd
import numpy as np
```

#### Importing superdataset

```python
superdataset = pd.read_csv('../../../../data/stage_1/superdataset_merge.csv')
superdataset.head()
```

#### Adding column for normalized data

```python
selective_df = superdataset[["County Code","Name","State","Population","Norm_Deaths","Premature death raw value", "Injury deaths raw value", "Opiod_Dispensing_Rate"]]
```

```python
logrithmic_df = superdataset[["County Code","Name","State","Population","Norm_Deaths","Premature death raw value", "Injury deaths raw value", "Opiod_Dispensing_Rate"]]
```

#### Normalize the variables for 100K population. 

```python
selective_df["Injury Death Rate"] = (selective_df["Injury deaths raw value"]/selective_df["Population"])*100000
selective_df["Premature Death Rate"] = (selective_df["Premature death raw value"]/selective_df["Population"])*100000
selective_df.head()
```

#### Calculating logarathimic values for variables

```python
#logrithmic_df['Opiod_Dispensing_Rate'] = np.log(logrithmic_df["Opiod_Dispensing_Rate"])
logrithmic_df['Injury Death Rate'] = np.log(logrithmic_df["Injury deaths raw value"])
logrithmic_df['Premature Death Rate'] = np.log(logrithmic_df["Premature death raw value"])
```

#### Removing negative and nan values

```python
#logrithmic_df = logrithmic_df[logrithmic_df['Opiod_Dispensing_Rate'] > 0]
logrithmic_df = logrithmic_df[logrithmic_df['Injury Death Rate'] > 0]
logrithmic_df = logrithmic_df[logrithmic_df['Premature Death Rate'] > 0]
logrithmic_df.head()
```

#### Get unique values of states

```python
states = selective_df["State"].unique()
```

```python
variables = ["Premature Death Rate","Injury Death Rate","Opiod_Dispensing_Rate"]
```

#### Use jupyter dash for creating dashboard

```python
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
from dash import Dash, dash_table
import plotly.graph_objects as go
from urllib.request import urlopen
import json

# Build App
app = JupyterDash(__name__)
app.layout = html.Div([
    
    # Flex container 
    html.Div([
        # Graph container
        html.Div([
            html.H1("Opioid Mortality Dashboard"),
            # State selector
            html.Div([
                html.Label([
                    "States",
                    dcc.Dropdown(
                        id='colorscale-dropdown', clearable=False,
                        value='NC', options=[
                        {'label': c, 'value': c}
                        for c in states
                    ])
                ]),
            ],style={'width': '50%', 'display': 'inline-block'}),
            # Variable selector
            html.Div([
                html.Label([
                    "Variables",
                    dcc.Dropdown(
                        id='variables-dropdown', clearable=False,
                        value='Premature Death Rate', options=[
                        {'label': k, 'value': k}
                        for k in variables
                    ])
                ])
            ],style={'width': '50%', 'display': 'inline-block'}),  
            #Radio button Linear and Log
            html.Div([
                    dcc.RadioItems(
                        id = 'states-radio',
                        options=[{'label': 'Linear', 'value': 'linear'},{'label': 'Log', 'value': 'log'}],
                        value='linear'
                    )
            ],style={'width': '50%', 'display': 'inline-block'}),
            #Radio button Linear and Log
            html.Div([
                    dcc.RadioItems(
                        id = 'variables-radio',
                        options=[{'label': 'Linear', 'value': 'linear'},{'label': 'Log', 'value': 'log'}],
                        value='linear'
                    )
            ],style={'width': '50%', 'display': 'inline-block'}), 
            
            html.Br(),
            html.Br(),
            #Radio button for Map or Scatter Plot
            html.Div([
                html.Label([
                    "Select the Output Graph:",
                    dcc.RadioItems(
                        id='toggle-radio',
                        options=[{'label': i, 'value': i} for i in ['Scatter Plot', 'Map']],
                        value='Scatter Plot'
                    )
                ])
            ],style={'width': '50%', 'display': 'inline-block'}),
                    
            dcc.Graph(id='graph')
        ]),
    ]),
    # Table container
    html.Div(id='table-container'),
])

# Define callback to update graph
@app.callback(
    Output('graph', 'figure'),
    [Input("colorscale-dropdown", "value"),
    Input("variables-dropdown", "value"), 
    Input("states-radio", "value"),
    Input("variables-radio", "value"),
    Input("toggle-radio", "value")]
)
def update_figure(state,variable,state_radio,variables_radio,toggle_radio):
    
    #Get all county in US state
    with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
        counties = json.load(response)
        
    if(state_radio == "linear" and variables_radio == "linear"):
        df = selective_df
    elif(state_radio == "log" or variables_radio == "log"):
        df = logrithmic_df
    else:
        df = logrithmic_df
        
    # Condition for scatter plot vs map
    if (toggle_radio == 'Scatter Plot'):
        fig = px.scatter( df.loc[(df['State']==state)], x= "Population", y="Norm_Deaths",
                size= variable, color= "State", 
                trendline="ols",opacity=0.5,
                size_max=20,color_discrete_sequence=px.colors.qualitative.Alphabet)
    else:
        fig = px.choropleth(
            df, geojson=counties, color= variable,
            locations='County Code', 
            projection="mercator", 
            range_color=[0, 6500])
        fig.update_geos(fitbounds="locations", visible=False)
        fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    return fig

@app.callback(
    Output('table-container', 'children'),
    [Input("colorscale-dropdown", "value"),
    Input("variables-dropdown", "value"), 
    Input("states-radio", "value"),
    Input("variables-radio", "value")]
)

def update_data(state,variable,state_radio,variables_radio):
    
    if(state_radio == "linear" and variables_radio == "linear"):
        df = selective_df.loc[(selective_df['State']==state)]
    elif(state_radio == "log" or variables_radio == "log"):
        df = logrithmic_df.loc[(logrithmic_df['State']==state)]
    else:
        df = logrithmic_df.loc[(logrithmic_df['State']==state)]

    return dash_table.DataTable(
        data=df.to_dict('records'),
        columns=[{'name': i, "id": i} for i in df.columns],
        page_size=10,
    )
# Run app and display result inline in the notebook
app.run_server(mode='inline')
```

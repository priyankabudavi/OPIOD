### Understand the distribution and measures of center for US opioid mortality for 2019 

* To understand the distribution i.e whether the data is continous or discrete, we first check if there are any infinity values,using the isinfite function.

```python
import numpy as np
srcd=superdf['Norm_Deaths']
x=np.isfinite(srcd)
x[~x]
```

* We, can see that there are no infite values in the opiod mortality rate.
* We can also analyse this visually by plotting a histogram.

```python
superdf['Norm_Deaths'].hist()
```

* From the above generated histogram we can see that the opiod mortality rate is <b> Discrete data</b>
* Also, we can see that the data is not uniformly distributed , so we use the measure of center as median to calculate the center.

```python
deathdf=superdf['Norm_Deaths']
median_deathdf=superdf['Norm_Deaths'].median()
median_deathdf
```

* Therefore the <b> Measure of center-median in our case is:</b> 9.0
* However, if we chose mean as a measure of center then,

```python
meandf=superdf['Norm_Deaths'].mean()
meandf
```

* Then the mean of the mortality rate would be <b>Measure of center:mean = 10.1459</b>

##### Here we codify the column Normalized Deaths (Number of Deaths per 100k Population - Norm_Deaths) and store it as a label column. (10 pts)
* This task was already performed in Stage-1 and we have stored Norm_Deaths as a column in our super_df

```python
superdf.columns
```

* We can see that superdf already has a column named Norm_Deaths which was generated by normalising the population per 100k

#### The codification is based on the quantile distribution of the Normalized deaths. i.e., Very Low (v_low) - [1.999 - 8.0],Low (low) - [8.0 - 11.0],High (high) - [11.0 - 16.0],Very High (v_high) - [16.0 - 64.0]

* To split the Normalised deaths to a quartile ranges and lable them we use the pandas cut operator.Which was referenced using,
https://mungingdata.com/pandas/dataframe-cut-category/

```python
superdf['quantile_distribution']=pd.cut(x=superdf['Norm_Deaths'],bins=[0.1,7.0,10.0,15.0,np.inf],
                                        labels=['Very_low','Low','High','Vey_high'])
superdf_x=superdf[['Norm_Deaths','quantile_distribution']]
superdf_x
```

* Here we, alter the quartile ranges so that all the values fall into somr quantile and none of the values are left out.

```python
superdf['quantile_distribution'].unique()
```

## Extra credit (additional 20 pts)
Develop the above approach as a Plotly widget for jupyter, where you can try out the variables dynamically in a simple dashboard. The variables can be in a dropdown.

```python
import plotly.graph_objects as go
from ipywidgets import widgets

#create a small datafrme including only few columns
newdf=superdf[['Total female population raw value','Total male population raw value','Opiod_Dispensing_Rate',
               'Population','Norm_Deaths','quantile_distribution']]
newdf.head()
```

```python
list1=['Total female population raw value','Total male population raw value','Opiod_Dispensing_Rate',
               'Population']
box2 = widgets.Dropdown(
    description='Mortality cause:',
    value='Population',
    options=list1
)
trace1 = go.Scatter(x=newdf["Norm_Deaths"], y=newdf["Population"],mode='markers')
g = go.FigureWidget(data=trace1,          
                    layout=go.Layout(      
                        title = 'Opiod mortality when compared with Population', 
                        xaxis_title = 'Norm_Deaths',         
                        yaxis_title = 'Population' ))
```

```python
def response(change):
    with g.batch_update():                       
        g.data[0].y = newdf[box2.value]
        g.layout.title = "opioid Mortality when comapred with "+box2.value  
        g.layout.xaxis.title = 'Norm_Deaths' 
        g.layout.yaxis.title = box2.value  

box2.observe(response, names="value")
```

```python
container2 = widgets.HBox([box2])
widgets.VBox([container2,g])
```
